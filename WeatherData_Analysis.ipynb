# Random Weather Data Analysis:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
weather_pd = pd.read_csv('random_weather_data.csv')
```

```python
weather_pd.head()
```

```python
# Create figure for Temperature vs. Latitude Plot:

# Stylize plot and create axes
plt.style.use('classic')
fig, ax = plt.subplots(figsize=(15,10))

# Create plot using color mapping for temperature:
weather_pd.plot(x='Temp (F)', y='Latitude', kind='scatter', ax=ax,
                s=50, fontsize=15, c='Temp (F)', cmap='coolwarm')

# Set Labels and axes limits
ax.set_title('Temperature vs. Latitude', fontsize=25, pad=15)
ax.set_xlabel('Temperature (F)', fontsize=15)
ax.set_ylabel('Latitude', fontsize=15)
ax.set_xlim(0,120)
ax.set_ylim(-90,90)
ax.set_yticks(np.arange(-90,90,20))


plt.grid()

plt.savefig('Figures\Temp vs Lat.png')
plt.show()
```

```python
# There are 5 points where the humidity exceeds 100%
weather_pd.loc[weather_pd['Humidity (%)'] > 100]
```

```python
# Should reset those 5 points to be 100% instead of over 100%
over_humidity = weather_pd.loc[weather_pd['Humidity (%)'] > 100].index
weather_pd.loc[over_humidity,'Humidity (%)'] = 100
weather_pd.loc[weather_pd['Humidity (%)'] > 100]
```

```python
# Create figure for Latitude vs. Humidity Plot:

# Stylize plot and create axes
plt.style.use('classic')
fig, ax = plt.subplots(figsize=(15,10))

# Create plot using color mapping for temperature:
weather_pd.plot(x='Latitude', y='Humidity (%)', kind='scatter', ax=ax,
                s=50, fontsize=15, c='Humidity (%)', cmap='coolwarm')

# Set Labels and axes limits
ax.set_title('Latitude vs. Humidity', fontsize=25, pad=15)
ax.set_xlabel('Latitude', fontsize=15)
ax.set_ylabel('Humidity (%)', fontsize=15)
ax.set_xlim(-90,90)
ax.set_ylim(0,105)
ax.set_yticks(np.arange(0,105,10))


plt.grid()

plt.savefig('Figures\Lat vs Humidity.png')
plt.show()
```

# Observable Trends:

### 1.) Maximum temperatures do not increase when the latitude ranges from -10 to 40.

- The hottest cities analyzed are around -10 and 40, and not exactly at the equator. However at a latitude of 40, the temperature can vary anywhere from 10 to 90 degrees fahrenheit, although 10 degrees is very rare.

### 2.) There is a very small portion of cities with low Humidity, but they do sometimes occur around -20 & 30 Latitude.

- From the entire dataset, there are only a handful of cities that have a humidity percentage of less than 20%. These cities are incredibly dry and have very high temperatures. Humidity around the equator and cities more than 30 from the equator are generally high around +80%.

### 3.)	There is an even distribution for cloudiness and wind speed when compared to the latitude.

- Most Cloudiness percentages are around either 0 to 10% or around 90% and most wind speeds are around 0 to 5 mph. There does not seem to be a significant correlation with these parameters and the latitude as the scatter plots are evenly distributed.
